```python
from matplotlib import cm
from matplotlib import pyplot
import math

def f(x, y):
   return x ** 2 + y ** 2

def gd(tx, ty):
  epoch = 30
  ita = 0.1
  gamma = 0.9
  solutionsX,solutionsY, scores = [], [], []
  ind = random.randrange(0,len(x)-1)
  solutionX = tx[ind]
  solutionY = ty[ind]
  w_look_aheadX=0
  w_look_aheadY=0
  #alpha = 0
  updateX=0
  updateY=0

  for i in range(epoch):

    gradX = 2*solutionX # 2*x
    gradY = 2*solutionY # 2*y
    
    for i in range(len(solutionsX)):
      w_look_aheadX += solutionsX[i] - gamma*updateX
      w_look_aheadY += solutionsY[i] - gamma*updateY

    updateX = gamma*updateX + ita*(gradX)
    updateY = gamma*updateY + ita*(gradY)

    solutionX = solutionX - updateX
    solutionY = solutionY - updateY
    solution_eval = solutionX**2 + solutionY**2 # x^2

    solutionsX.append(solutionX)
    solutionsY.append(solutionY)
    scores.append(solution_eval)

    print('>%d f(%s, %s) = %.5f' % (i, solutionX,solutionY, solution_eval))

    if(solution_eval.round(4) == 0.0000):
      print("found")
      break
  return [solutionsX,solutionsY,scores]

newX,newY, score = gd(x, y)

X, Y = np.meshgrid(newX, newY)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap=cm.coolwarm)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.show()
```
