```python
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from keras.datasets import mnist
```

```python
import numpy as np
(train_X, train_y), (test_X, test_y) = mnist.load_data()

num_images = 100
idx = np.random.choice(train_X.shape[0], num_images, replace=False)
train_X = train_X[idx]
train_y = train_y[idx]

num_images = 20
idx = np.random.choice(test_X.shape[0], num_images, replace=False)
test_X = test_X[idx]
test_y = test_y[idx]
```

```python
from matplotlib import pyplot
for i in range(9):  
  pyplot.subplot(330 + 1 + i)
  pyplot.imshow(train_X[i], cmap=pyplot.get_cmap('gray'))
  pyplot.show()
```

```python
train_X = tf.pad(train_X, [[0, 0], [2,2], [2,2]])/255
test_X = tf.pad(test_X, [[0, 0], [2,2], [2,2]])/255
train_X = tf.expand_dims(train_X, axis=3, name=None)
test_X = tf.expand_dims(test_X, axis=3, name=None)
train_X = tf.repeat(train_X, 3, axis=3)
test_X = tf.repeat(test_X, 3, axis=3)
```

```python

from tensorflow.keras import layers, models, losses

alexnet_model = models.Sequential()

alexnet_model.add(layers.experimental.preprocessing.Resizing(224, 224, interpolation="bilinear", input_shape=train_X.shape[1:]))


alexnet_model.add(layers.Conv2D(96, 11, strides=4, padding='same'))
alexnet_model.add(layers.Lambda(tf.nn.local_response_normalization))
alexnet_model.add(layers.Activation('relu'))
alexnet_model.add(layers.MaxPooling2D(3, strides=2))

alexnet_model.add(layers.Conv2D(256, 5, strides=4, padding='same'))
alexnet_model.add(layers.Lambda(tf.nn.local_response_normalization))
alexnet_model.add(layers.Activation('relu'))
alexnet_model.add(layers.MaxPooling2D(3, strides=2))

alexnet_model.add(layers.Conv2D(384, 3, strides=4, padding='same'))
alexnet_model.add(layers.Activation('relu'))

alexnet_model.add(layers.Conv2D(384, 3, strides=4, padding='same'))
alexnet_model.add(layers.Activation('relu'))

alexnet_model.add(layers.Conv2D(256, 3, strides=4, padding='same'))
alexnet_model.add(layers.Activation('relu'))

alexnet_model.add(layers.Flatten())
alexnet_model.add(layers.Dense(4096, activation='relu'))
alexnet_model.add(layers.Dropout(0.5))

alexnet_model.add(layers.Dense(4096, activation='relu'))
alexnet_model.add(layers.Dropout(0.5))

alexnet_model.add(layers.Dense(10, activation='softmax'))
alexnet_model.summary()
     
```

```python
alexnet_model.compile(optimizer='adam', loss=losses.sparse_categorical_crossentropy, metrics=['accuracy'])
```

```python
hist_alex = alexnet_model.fit(train_X, train_y, validation_data=(test_X, test_y), batch_size=64, epochs=10)
```

```python
lenet5_model = models.Sequential()

lenet5_model.add(layers.experimental.preprocessing.Resizing(224, 224, interpolation="bilinear", input_shape=train_X.shape[1:]))


lenet5_model.add(layers.Conv2D(6, 11, strides=4, padding='same'))
lenet5_model.add(layers.Lambda(tf.nn.local_response_normalization))
lenet5_model.add(layers.Activation('relu'))
lenet5_model.add(layers.MaxPooling2D(2, strides=2))

lenet5_model.add(layers.Conv2D(16, 5, strides=4, padding='same'))
lenet5_model.add(layers.Lambda(tf.nn.local_response_normalization))
lenet5_model.add(layers.Activation('relu'))
lenet5_model.add(layers.MaxPooling2D(2, strides=2))

lenet5_model.add(layers.Flatten())
lenet5_model.add(layers.Dense(120, activation='relu'))
lenet5_model.add(layers.Dropout(0.5))

lenet5_model.add(layers.Dense(84, activation='relu'))
lenet5_model.add(layers.Dropout(0.5))

lenet5_model.add(layers.Dense(10, activation='softmax'))
lenet5_model.summary()
```

```python
lenet5_model.compile(optimizer='adam', loss=losses.sparse_categorical_crossentropy, metrics=['accuracy'])
hist_lenet = lenet5_model.fit(train_X, train_y, validation_data=(test_X, test_y), batch_size=64, epochs=10)
```

```python
vgg16_model = Sequential()

vgg16_model.add(layers.experimental.preprocessing.Resizing(224, 224, interpolation="bilinear", input_shape=train_X.shape[1:]))

vgg16_model.add(layers.Conv2D(64, 11, strides=4, padding='same'))
vgg16_model.add(layers.Lambda(tf.nn.local_response_normalization))
vgg16_model.add(layers.Activation('relu'))
vgg16_model.add(layers.MaxPooling2D(3, strides=2))

vgg16_model.add(layers.Conv2D(128, 5, strides=4, padding='same'))
vgg16_model.add(layers.Conv2D(128, 5, strides=4, padding='same'))
vgg16_model.add(layers.Lambda(tf.nn.local_response_normalization))
vgg16_model.add(layers.Activation('relu'))
vgg16_model.add(layers.MaxPooling2D(2, strides=2))


vgg16_model.add(layers.Conv2D(256, 3, strides=4, padding='same'))
vgg16_model.add(layers.Conv2D(256, 3, strides=4, padding='same'))
vgg16_model.add(layers.Conv2D(256, 3, strides=4, padding='same'))
vgg16_model.add(layers.Activation('relu'))

vgg16_model.add(layers.Conv2D(512, 3, strides=4, padding='same'))
vgg16_model.add(layers.Conv2D(512, 3, strides=4, padding='same'))
vgg16_model.add(layers.Conv2D(512, 3, strides=4, padding='same'))
vgg16_model.add(layers.Conv2D(512, 3, strides=4, padding='same'))
vgg16_model.add(layers.Conv2D(512, 3, strides=4, padding='same'))
vgg16_model.add(layers.Conv2D(512, 3, strides=4, padding='same'))
vgg16_model.add(layers.Activation('relu'))

vgg16_model.add(layers.Flatten())
vgg16_model.add(layers.Dense(4096, activation='relu'))
vgg16_model.add(layers.Dropout(0.5))

vgg16_model.add(layers.Dense(4096, activation='relu'))
vgg16_model.add(layers.Dropout(0.5))

vgg16_model.add(layers.Dense(10, activation='softmax'))
vgg16_model.summary()
     
```

```python
vgg16_model.compile(optimizer='adam', loss=losses.sparse_categorical_crossentropy, metrics=['accuracy'])
hist_vgg16 = vgg16_model.fit(train_X, train_y, validation_data=(test_X, test_y),batch_size=64, epochs=10)
```

```python
from matplotlib import pyplot as plt
plt.plot(hist_alex.history['accuracy'])
plt.plot(hist_alex.history['val_accuracy'])

plt.title('Alexnet model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
plt.plot(hist_lenet.history['accuracy'])
plt.plot(hist_lenet.history['val_accuracy'])

plt.title('Lenet5 model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
plt.plot(hist_vgg16.history['accuracy'])
plt.plot(hist_vgg16.history['val_accuracy'])

plt.title('VGG16 model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
